```python
from config import *
from collections import Counter

import numpy as np
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.mpl_style = 'default'
```

```python
df_train = pd.read_csv(PROCESSINGTEXT_DIR + '/df_train_processed.csv', encoding="ISO-8859-1")
df_test = pd.read_csv(PROCESSINGTEXT_DIR + '/df_test_processed.csv', encoding="ISO-8859-1")
df_solution = pd.read_csv(PROCESSINGTEXT_DIR + '/solution.csv', encoding="ISO-8859-1")

df_test['relevance'] = df_solution['relevance']
df_test['Usage'] = df_solution['Usage']
df_test = df_test[df_test['Usage'] != 'Ignored']
df_test = df_test.drop(['Usage'],axis=1)

df_train = df_train.sort_values('search_term')
train_count = Counter(df_train['search_term'])
df_test = df_test.sort_values('search_term')
test_count= Counter(df_test['search_term'])
```

```python
df_test[:50]
```

```python
df_train['query freq'] = df_train['search_term'].map(lambda x:train_count[x])
df_test['query freq'] = df_test['search_term'].map(lambda x:test_count[x])


df_train_pair_wise = df_train[df_train['query freq']>4]
df_test_pair_wise = df_test[df_test['query freq']>4]
```
